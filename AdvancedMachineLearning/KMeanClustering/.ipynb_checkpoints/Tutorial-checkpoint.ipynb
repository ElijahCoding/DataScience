### Do clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = data[['x', 'y']]
y = data['cluster']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
data[['x', 'y']] = X
```

```python
kmeans = KMeans(n_clusters=2, random_state=3)
kmeans.fit(X)
```

```python
center = kmeans.cluster_centers_
center
```

```python
plt.scatter(data['x'], data['y'], c = kmeans.labels_, cmap = 'viridis')
for i, point in enumerate(center):
    plt.plot(center[i][0], center[i][1], '*r--', linewidth=2, markersize=18)
plt.show()
```

```python
plt.scatter(data['x'], data['y'], c = data['cluster'], cmap = 'viridis')
plt.show()
```

### Choose K

Step 1 - Pick K random points as cluster centers called centroids. <br>
Step 2 - Assign each x_i to nearest cluster by calculating its distance to each centroid. <br>
Step 3 - Find new cluster center by taking the average of the assigned points. <br>
Step 4 - Repeat Step 2 and 3 until none of the cluster assignments change. <br>

```python
SSE = []
index = range(1, 10)

for i in index:
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(X)
    SSE.append(kmeans.inertia_)
    print(kmeans.inertia_)
```

```python
plt.plot(index, SSE)
plt.xlabel('K')
plt.ylabel('SEE')
plt.show()
```

```python
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
iris.target_names
```

```python
SSE = []

index = range(1, 10)

for i in index:
    kmeans = KMeans(n_clusters=i, random_state=4)
    kmeans.fit(X)
    SSE.append(kmeans.inertia_)
    print(kmeans.inertia_)
```

```python
plt.plot(index, SSE)
plt.xlabel('K')
plt.ylabel('SSE')
plt.show()
```
